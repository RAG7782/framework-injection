## 3. A régua de bolso — leia o seu próprio FI
Para cada cláusula que saiu acima, classifique:

| Cláusula | Induz (priming) | Constrange (enforcement) | Deveria ser? |
|---|---|---|---|
| `[PAPEL]` | ✅ | | priming ok |
| `[MÉTODO]` | ✅ | | priming ok |
| `[META-RACIOCÍNIO]` | ✅ | | priming ok |
| `[VERIFICAÇÃO]` (ASSERT executável) | | ✅ | **enforcement** — só vale se dispara teste real |

> **Pergunta-régua:** "isto induz ou constrange? E qual dos dois *deveria*?"  
> Cláusula que induz onde deveria constranger = **dívida**.  
> Maturidade do FI = % de âncoras críticas que escalaram de priming a enforcement.
